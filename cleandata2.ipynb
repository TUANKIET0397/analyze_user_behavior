```python
import pandas as pd
import numpy as np
from sklearn.utils import resample

# Đọc dữ liệu
df = pd.read_csv('shopping_trends_updated.csv')
df.head()
```

```python
# Định nghĩa các khoảng tuổi và nhãn tương ứng
bins = [18, 30, 45, 60, 100]
labels = ['Young Adult', 'Middle Aged', 'Senior', 'Elderly']

df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Kết quả chia nhóm
print(df['Age Group'].value_counts())
```

```python
# Phân phối các mùa
print("Trước khi cân bằng:\n", df['Season'].value_counts())

# Tách các nhóm theo mùa
df_spring = df[df['Season'] == 'Spring']
df_fall = df[df['Season'] == 'Fall']
df_winter = df[df['Season'] == 'Winter']
df_summer = df[df['Season'] == 'Summer']

# Lấy số lượng mẫu thấp nhất của một mùa để làm chuẩn 
min_samples = min(len(df_spring), len(df_fall), len(df_winter), len(df_summer))

# Lấy mẫu lại cho bằng nhau
df_spring_resampled = resample(df_spring, replace=False, n_samples=min_samples, random_state=42)
df_fall_resampled = resample(df_fall, replace=False, n_samples=min_samples, random_state=42)
df_winter_resampled = resample(df_winter, replace=False, n_samples=min_samples, random_state=42)
df_summer_resampled = resample(df_summer, replace=False, n_samples=min_samples, random_state=42)

# Gộp lại thành data mới đã cân bằng
df_balanced = pd.concat([df_spring_resampled, df_fall_resampled, df_winter_resampled, df_summer_resampled])

print("\nSau khi cân bằng:\n", df_balanced['Season'].value_counts())
```

```python
# Chọn các cột cần One-Hot Encoding gồm 'Age Group' vừa tạo và các cột khác
categorical_cols = ['Age Group', 'Gender', 'Item Purchased', 'Category', 'Location', 'Season', 'Subscription Status']

# Thực hiện One-Hot Encoding bằng pandas
df_final = pd.get_dummies(df_balanced, columns=categorical_cols)

# Hiển thị kết quả cuối cùng
df_final.head()
```

```python
# Lưu data đã cân bằng và file csv
df_final.to_csv('shopping_trends_cleaned_encoded.csv', index=False)

print("Đã xuất file thành công: shopping_trends_cleaned_encoded.csv")
```
